```python


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/Data set/BollywoodMovieDetail.csv')
data.head()
```

```python
for col in data.columns:
  null_value=np.mean(data[col].isnull())
  print(f'{col} :- {null_value*100}%')
```

```python
for col in data.columns:
  print(f'{col} :- {data[col].isnull().sum()}')
```

```python
data1=data.dropna(subset=['genre','actors','directors','sequel',]) # remove rows having least null values
print(data1.columns)
for col in data1.columns:
    print(f'{col} :- {data1[col].isnull().sum()}')
```

```python
data1['releaseDate']=pd.Categorical(data1['releaseDate'])   # convert string elemt to categorical data
top_value1=data['releaseDate'].describe()['top'] # calculate mode for the column data
data1['releaseDate']=data1['releaseDate'].fillna(top_value1) # fill nan value with mode value

data1['writers']=pd.Categorical(data1['writers'])  # convert string elemt to categorical data
top_value2=data['writers'].describe()['top']  # calculate mode for the column data
data1['writers']=data1['writers'].fillna(top_value2) # fill nan value with mode value


for col in data1.columns:
  print(f'{col} :- {data1[col].isnull().sum()}')
```

```python
data1.head()
```

```python
# Now check for outliers value in numerical data

sns.boxplot(data1['hitFlop'])
```

```python
q3=data1['hitFlop'].quantile(0.75)
q1=data1['hitFlop'].quantile(0.25)
IQR=q3-q1
#print(IQR)
upper=q3+1.5*IQR
lower=q1-1.5*IQR
```

```python
# Now data is clean means no Nan value & outliers

data1=  data1[(data1['hitFlop']<upper) & (data1['hitFlop']>lower)]
sns.boxplot(data1['hitFlop'])
```

```python
# Create a list of all genres
all_genres = data['genre'].str.cat(sep=' | ').split(' | ')
print(all_genres)
```

```python
  # Create a DataFrame with genre names and their counts
genre_counts = pd.DataFrame(all_genres,columns=['genre']).value_counts().reset_index()
genre_counts.columns=['genre','MovieCount']
genre_counts=genre_counts.head()
genre_counts
```

```python
# Get the top 5 genres
top_genres = genre_counts['genre'].tolist()
print(top_genres)
```

```python
# Create a DataFrame to store the percentage of movies in each genre for each year

# new command

genre_trends = pd.DataFrame(index=data['releaseYear'].sort_values().unique(), columns=top_genres) # new command
genre_trends.head()
```

```python
# Fill the DataFrame
for genre in top_genres:
    for year in genre_trends.index:
        # data frame for particaular year having all genre movie
        year_movies = data[data['releaseYear'] == year]
        # data frame for particaular year having particular genre movie
        genre_movies = year_movies[year_movies['genre'].str.contains(genre, na=False)]
        genre_trends.loc[year, genre] = len(genre_movies) / len(year_movies) * 100
```

```python
#print(genre_trends.index)
year=[2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011,2012, 2013, 2014]
plt.plot(year,genre_trends['Drama'])
```

```python
# Plot the genre trends over time
plt.figure(figsize=(10, 4))
for genre in top_genres:
   plt.plot(genre_trends.index, genre_trends[genre], label=genre) # new command

plt.title('Genre Trends Over Time (Top 5 Genres)')
plt.xlabel('Year')
plt.ylabel('Percentage of Movies (%)')
plt.legend(title='Genre')
plt.grid()
plt.show()
```
